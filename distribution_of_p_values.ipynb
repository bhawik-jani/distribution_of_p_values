```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

## Distribution of p-values Under True Null Hypothesis

```python
# H0: true mean is 0
np.random.seed(42)

n_simulations = 100000
n = 30  # Sample size
mu_true = 0  # Population mean
sigma = 1  # Population standard deviation

p_values = []

for _ in range(n_simulations):
    sample = np.random.normal(mu_true, sigma, n)    # Generate a sample
    mu_sample = sample.mean()
    t_stat = (mu_sample - 0) / (sigma / (n ** 0.5))    # Compute t-statistic
    p_val = 2 * (1 - stats.t.cdf(abs(t_stat), df=n-1))    # Perform t-test
    p_values.append(p_val)

# Plot the distribution of p-values
plt.hist(p_values, bins=50, edgecolor='black', density=True)
plt.xlabel("P-Value")
plt.ylabel("Density")
plt.title("Distribution of P-Values Under H0")
plt.show()
```

## Distribution of p-values Under False Null Hypothesis

```python
# H0: true mean is 0
np.random.seed(42)

mu_true = 0.2  # True mean different from H0

p_values = []

for _ in range(n_simulations):
    sample = np.random.normal(mu_true, sigma, n)    # Generate a sample
    mu_sample = sample.mean()
    t_stat = (mu_sample - 0) / (sigma / (n ** 0.5))    # Compute t-statistic
    p_val = 2 * (1 - stats.t.cdf(abs(t_stat), df=n-1))    # Perform t-test
    p_values.append(p_val)

# Plot the distribution of p-values
plt.hist(p_values, bins=50, edgecolor='black', density=True)
plt.xlabel("P-Value")
plt.ylabel("Density")
plt.title("Distribution of P-Values Under H0")
plt.show()
```
